```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Display first few rows
titanic.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['age'].dropna(), bins=10, kde=True)
plt.title("Distribution of Passenger Ages")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="class", data=titanic, hue="class", palette="Set2", legend=True)

plt.title("Passenger Count by Class")
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(6,4))
sns.barplot(x="sex", y="survived", data=titanic,hue="sex" , palette="coolwarm")
plt.title("Survival Rate by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="class", y="age", data=titanic, hue="class" ,palette="pastel")
plt.title("Age Distribution Across Classes")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(
    x="age",
    y="survived",
    data=titanic,
    hue="sex",        # color by gender
    style="class",    # shape by passenger class
    palette="coolwarm",
    alpha=0.7
)
plt.title("Scatter Plot: Age vs Survival (Colored by Gender, Styled by Class)")
plt.xlabel("Age")
plt.ylabel("Survival (0 = No, 1 = Yes)")
plt.show()
```

```python
# Compute average fare per class
fare_by_class = titanic.groupby('class')['fare'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(
    x='class',
    y='fare',
    data=fare_by_class,
    marker='o',        # show points on line
    color='blue'
)
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="class", y="age", hue="sex", data=titanic, split=True, palette="muted")
plt.title("Age Distribution by Class and Gender")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Stripplot of Age by Class
plt.figure(figsize=(8,5))
sns.stripplot(x="class", y="age", data=titanic, hue="class",jitter=True, palette="Set2")
plt.title("Stripplot of Age across Passenger Classes")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x="class", y="age", data=titanic,hue="class", palette="Set1")
plt.title("Swarmplot of Age across Passenger Classes")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Data")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
titanic_numeric = titanic.select_dtypes(include=["float64", "int64"])

# Compute correlation matrix
corr = titanic_numeric.corr()

# Plot clustermap
sns.clustermap(corr, annot=True, cmap="coolwarm", figsize=(8,6))
plt.suptitle("Cluster Map of Titanic Numerical Feature Correlations", y=1.02)
plt.show()
```

```python
sns.pairplot(titanic[["age", "fare", "survived", "pclass"]].dropna(), hue="survived")
plt.suptitle("Pairplot of Selected Titanic Features", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Create figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Histogram of Age
sns.histplot(titanic["age"], kde=True, bins=30, ax=axes[0,0], color="skyblue")
axes[0,0].set_title("Age Distribution")

# 2. Boxplot of Fare by Class
sns.boxplot(x="class", y="fare", data=titanic, ax=axes[0,1], palette="Set2")
axes[0,1].set_title("Fare by Class")

# 3. Countplot of Gender
sns.countplot(x="sex", data=titanic, ax=axes[1,0], palette="pastel")
axes[1,0].set_title("Count of Male vs Female")

# 4. Scatter plot Age vs Fare
sns.scatterplot(x="age", y="fare", data=titanic, hue="class", ax=axes[1,1], palette="Set1")
axes[1,1].set_title("Age vs Fare (by Class)")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(titanic['age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution (Matplotlib)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
survival_counts = titanic['survived'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(survival_counts, labels=["Not Survived","Survived"], autopct='%1.1f%%', colors=["lightcoral","lightgreen"])
plt.title("Survival Distribution (Pie Chart)")
plt.show()
```

```python
avg_fare = titanic.groupby("class")["fare"].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_fare.index, avg_fare.values, color=["gold","silver","brown"])
plt.title("Average Fare by Class")
plt.xlabel("Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(titanic["age"], titanic["fare"], "o", markersize=4, alpha=0.5)
plt.title("Fare vs Age (Scatter/Line Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
