```python
df=pd.read_csv("Automobile - Automobile.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
#our horsepower coloumn has many missing values so delete that coloumn
df=df.drop(columns=["horsepower"])
```

```python
df.info()
```

```python
df.head(10)
```

```python
# fill with missing values
#acceleration has missing values ,cyclinders,weight
df['cylinders'].fillna(df['cylinders'].median(),inplace=True) #inplace means wherever nan is there it goes to that point and fills
df['weight'].fillna(df['weight'].median(),inplace=True)
df.head(10)
```

```python
nc=["mpg","displacement","weight","acceleration"]
SC1=StandardScaler()
df[nc]=SC1.fit_transform(df[nc])
df.head(10)
```

```python
SC2=MinMaxScaler()
df[nc]=SC2.fit_transform(df[nc])
df.head(10)
```

```python
df.to_csv("automobile.csv",index="False")
```
